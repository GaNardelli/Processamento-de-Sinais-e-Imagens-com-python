```python
#Importando Bibliotecas
import math
import numpy as np
import matplotlib.pyplot as plt
import random 
from scipy import signal
from scipy.fft import fft, fftfreq, fftshift
```

## a-) Crie o sinal x t = 35 cos 2π10t + 60cos(2π150t) + ruído aleatório (amplitude 15)
► Deixe o sinal com média 0

► Tempo de 0 a 1s; Fs = 2000 Hz

```python
# Sinal
t = np.arange(0,1.000,1/2000) # Frequência 2000Hz
xt = 35*np.cos(2*np.pi*10*t) + 60*np.cos(2*np.pi*150*t)
plt.rcParams['figure.figsize']=(14,6)
plt.plot(t,xt,'g-')
plt.title('Sinal Original')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')

# Sinal + Ruído
xr = xt + 15*np.random.random((1,len(xt)))[0]
plt.figure()
plt.rcParams['figure.figsize']=(14,6)
plt.plot(t,xr,'b-')
plt.title('Sinal Original + Ruído')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')

# Sinal + Ruído e média=0
media = sum(xr)/len(xr)
x0 = []
for i in range (0,len(xr)):
   x0.append(xr[i]-media)

print(media)
plt.figure()
plt.rcParams['figure.figsize']=(14,6)
plt.plot(t,xr,'r-')
plt.title('Sinal Original + Ruído (média=0)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
```

## b-) Projete um filtro passa-baixas FIR com n = 121, flp = 0,1, usando a janela “blackman”

```python
n= 101
flp=0.1 
lp=signal.firwin(n, flp,window='blackman')
```

## c-) Crie o sinal y(t), aplicando o filtro criado em b) no sinal x(t), e faça os gráficos (separados) x e y

```python
y1=signal.filtfilt(lp, 1, x0)

# Sinal + Filtro
plt.figure()
plt.rcParams['figure.figsize']=(14,6)
plt.plot(y1, 'g-')
plt.title('Sinal Original com filtro passa-baixa')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')

# Sinal + Ruído + Filtro
plt.figure()
plt.plot(x0, 'r-')
plt.rcParams['figure.figsize']=(14,6)
plt.title('Sinal Original + Ruído com filtro passa-baixa')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')


# Sinal + Ruído + Filtro e media=0
plt.figure()
plt.rcParams['figure.figsize']=(14,6)
plt.plot(y1,'y-', linewidth= 4.5)
plt.plot(x0, 'b-')
plt.title('Sinal Original + Ruído com filtro passa-baixa (média=0)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
```
