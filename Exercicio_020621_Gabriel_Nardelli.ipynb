# Exercício 4

```python

def curva_gaussiana(x, amplitude, mean, stddev):
    return amplitude * np.exp(-np.power(x - mean, 2) / (2 * np.power(stddev, 2)))

# Gerando uma lista aleatória 

l = [rnd.randint(0,100) for i in range(150)] # o módulo gauss gera valores em uma distribuição gaussiana com (média, desvio padrão)

# Cálculo da Média e Desvio padrão

media = statistics.mean(l)
desvpad = statistics.stdev(l)

# Mostrando o Histograma criado

#Plot
plt.figure(figsize=[10,8])
plt.xlabel('Valores')
plt.grid(axis='y', alpha=0.75)
plt.xticks()
plt.yticks()
plt.ylabel('Frequência')
plt.title('Histograma e Curva Gaussiana')
plt.hist(l, bins=16, color = 'blue') #plotando o histograma
x = np.linspace(0, 100, 1000)
y = curva_gaussiana(x, 12, media, desvpad)
plt.plot(x,y, color = 'black')
plt.show()
```

# Experimento de Amostragem

Primeiramente precisamos criar os sinais,  utilizamos a função **math.cos** para que o sinal seja senoidal e utilizamos do principio matematico onde **f(x) = Acos(wt + fi)**.

Fizemos então a plotagem individual de cada sinal primeira mente com Fs = 10 amostras/s e em seguida Fs = 20 amostras/s. E comparamos os resultados dos sinais 1 e 2.

```python
# Sinais cossenoidais
time = [i for i in np.arange(0, 1, 0.01)] # intervalos de 0 a 1 com espaçamento de 0.01

f1 = 3 # frequencia sinal 1
f2 = 7 # frequencia sinal 2
x1 = [math.cos(2*math.pi*f1*n) for n in np.arange(0, 1, 0.01)]
x2 = [2*math.cos(2*math.pi*f2*n) for n in np.arange(0, 1, 0.01)]

# Amostragem para Fs = 10 amostras/s
n = [j for j in np.arange(0, 1, 0.1)]

xa1 = [math.cos(2*math.pi*f1*n) for n in np.arange(0, 1, 0.1)]
xa2 = [2*math.cos(2*math.pi*f2*n) for n in np.arange(0, 1, 0.1)]
```

```python
# Plotagem dos resultados

plt.subplot(7,1,1)
plt.title("Sinais de amostragem para Fs=10 amostras por segundo", fontsize=10)
plt.plot(time, x1, 'c', linewidth=1)

plt.subplot(7,1,2)
plt.stem(n, xa1, 'c')

plt.subplot(7,1,3)
plt.plot(time, x1, 'c', linewidth=1)
plt.stem(n, xa1, '')

plt.subplot(7,1,4)
plt.plot(time, x2, 'c', linewidth=1)

plt.subplot(7,1,5)
plt.stem(n, xa2, 'c')

plt.subplot(7,1,6)
plt.plot(time, x2, 'c',linewidth=1)
plt.stem(n, xa2, 'c')

plt.subplot(7,1,7)
plt.plot(time, x1, 'c',linewidth=1)
plt.stem(n, xa1, 'c')
plt.plot(time, x2, 'c',linewidth=1)
plt.stem(n, xa2, 'c')
```

```python
# Fs = 20 amostras/s

n = [j for j in np.arange(0, 1, 0.05)]

xa1 = [math.cos(2*math.pi*f1*n) for n in np.arange(0, 1, 0.05)]
xa2 = [2*math.cos(2*math.pi*f2*n) for n in np.arange(0, 1, 0.05)]

# Plotagem dos resultados

plt.subplot(7,1,1)
plt.title("Sinais de amostragem para Fs=20 amostras por segundo", fontsize=10)
plt.plot(time, x1, 'c', linewidth=1)

plt.subplot(7,1,2)
plt.stem(n, xa1, 'c')

plt.subplot(7,1,3)
plt.plot(time, x1, 'c', linewidth=1)
plt.stem(n, xa1, '')

plt.subplot(7,1,4)
plt.plot(time, x2, 'c', linewidth=1)

plt.subplot(7,1,5)
plt.stem(n, xa2, 'c')

plt.subplot(7,1,6)
plt.plot(time, x2, 'c',linewidth=1)
plt.stem(n, xa2, 'c')

plt.subplot(7,1,7)
plt.plot(time, x1, 'c',linewidth=1)
plt.stem(n, xa1, 'c')
plt.plot(time, x2, 'c',linewidth=1)
plt.stem(n, xa2, 'c')
```

## Discussão 

Percebemos logo de início uma grande diferença entre Fs = 10 amostras/s e Fs = 20 amostras/s, no primeiro caso temos um gráfico onde existe um menor numero de pontos visto que 1 amostra seria gerada a cada 0.1s já no segundo caso, 1 amostra é gerada a cada 0.05s, assim fica clara a diferença entre os dois gráficos.

Podemos ver também a diferença de amplitude presente entre os sinais **xa1** e **xa2** no ultimo gráfico onde existe a comparacção entre ambos. 

# Experimento de Ruído e Convolução

```python
# Vetor controle de tempo 

time = [i for i in np.arange(0, 1, 0.001)] # intervalos de 0 a 1 com espaçamento de 0.01

# Criação do sinal 

signal = [9*math.cos(2*math.pi*3*t) + 3*math.cos(2*math.pi*25*t) for t in time]
```

```python
# Visualização do Sinal e do Sinal com ruído
signal_r = []
for i in signal: 
  signal_r.append(i * 2 * rnd.random()) # ruído com aplitude 2 por numeros aleatorios

signal_r2 = [9*math.cos(2*math.pi*3*t) + 3*math.cos(2*math.pi*25*t) + 2*math.sin(math.pi*t) for t in time] # ruido utilizando a função sen

plt.plot(time, signal_r, label = 'Sinal com ruído aleatório')
plt.plot(time, signal_r2, label = 'Sinal com ruído senoidal')
plt.plot(time, signal, label = 'Sinal Limpo')
plt.legend()
plt.title("Coparacação entre sinais")
plt.xlabel('Tempo')
plt.show()
```

```python
# Criação do sinal h(t')

time_line = [i for i in np.arange(0, 101, 1)]

h = [math.exp((-1)*math.pow((t - 50),2)/(600)) for t in time_line]

# Visualização do sinal

plt.title("Sinal h(t')")
plt.xlabel("Tempo")
plt.plot(time_line, h, label = "Sinal h(t')")
plt.legend()
plt.show()
```

```python
# Convolução dos sinais

L=1000+101-1
l=[k for k in np.arange(0,L,1)]
convolved = np.convolve(signal_r2, h) # convolução do sinal com ruido senoidal
convolved_2 = np.convolve(signal_r,h) # convolução do sinal com ruido aleatorio

sub1 = plt.subplot(1,3,1)
plt.title('Sinal com ruído', fontsize = 12)
plt.plot(time, signal_r2)
plt.setp(sub1.get_yticklabels(), fontsize=5)
plt.setp(sub1.get_xticklabels(), fontsize=10)

sub2 = plt.subplot(1,3,2)
plt.title("Sinal h(t')", fontsize = 12)
plt.plot(time_line,h)
plt.setp(sub2.get_yticklabels(), fontsize=5)
plt.setp(sub2.get_xticklabels(), fontsize=10)

sub3 = plt.subplot(1,3,3)
plt.title('Sinal concolucionado', fontsize = 12)
plt.plot(l, convolved)
plt.setp(sub3.get_yticklabels(), fontsize=5)
plt.setp(sub3.get_xticklabels(), fontsize=10)
```

## Discussão 

Podemos perceber que ao fazer a convolução do sinal ruidoso com o sinal **h(t')** temos uma ótima resposta e assim conseguimos filtrar o sinal para que sejamos capazes de observar e estudar o comportamento.

Entretanto é necessário prestar atenção que os valores e escalas utilizados para gerar o primeiro sinal podem ser alterados devido a convolução, assim, não podemos nos basear inteiramente no sinal convolucionado.
