# Experimento transformada de Fourier 

As células anteriores aos exercícios são as funções criadas para auxiliar a realização dos mesmos.

```python
import random as rnd
import os # funções do sistema
import matplotlib.pyplot as plt
import numpy as np
from scipy import fft, signal

def cria_arquivo(nome_arquivo, quantidade_numeros):
  with open(f'{nome_arquivo}.dat', 'w') as f:
    for i in range(quantidade_numeros):
      f.write(f'{rnd.randrange(0,3)}\n')

def le_arquivo(arquivo, cria = 0):
  lista = []
  with open(os.path.abspath(arquivo), 'r') as f:
    for i,line in enumerate(f):
      lista.append(float(f.readline().strip()))

  grafico = cria_grafico(cria,[i for i in range(len(lista))], lista)

  return lista

def cria_grafico(cria,x,y):
  if cria == 0:
    pass
  else:
    plt.plot(x,y)
    plt.grid()
    plt.show()


def transformada(sinal, fs):
  transformada = fft(sinal)
  f,t, sxx = signal.spectrogram(transformada, fs)
  plt.pcolormesh(t,f,sxx,shading='gouraud', cmap='PuOr')
  plt.xlabel('Tempo (s)')
  plt.ylabel('Frequências (Hz)')
  plt.colorbar()
  plt.show()
```

```python
nome = 'arquivo1'
cria_arquivo(nome, 256)
valores = le_arquivo(f'{nome}.dat',1)
```

```python
spindle_data = open("spindle.dat", 'r')
spindle_signal = spindle_data.readlines()
print(spindle_signal)
tempo_spindle = np.linspace(0, 10, len(spindle_signal))
plt.plot(tempo_spindle, spindle_signal)
ams_spindle = len(spindle_signal)
freq_spindle = np.fft.fftfreq(len(spindle_signal))*ams_spindle
fft_spindle = np.fft.fft(spindle_signal)

plt.plot(freq_spindle[0:512], np.abs(fft_spindle[0:512]))
plt.grid(True)
plt.xlim(0, 35)
plt.title("FFT de spindle")
plt.xlabel("Frequencia (Hz)")
plt.ylabel("Amplitude")

fs = 256
f, t , Sxx = signal.spectrogram(fft_spindle, fs)
plt.pcolormesh(t, f, Sxx, cmap = 'autumn')
plt.colorbar()
plt.ylabel("Frequencia (Hz)")
plt.xlabel("Time")
plt.xlim(0.5, 0.95)
plt.ylim(-128, 50)
```

## Discussão

Podemos analisar as frequências ao decorrer do tempo e como ocorreu sua variação em freqencias mais altas, temos também na região inferior uma clara variação nas frequências geradas após a utilização das tranformadas para o sinal, temos assim, que apartir da transformadad rápida de fourier somos capazes de gerar um espectrograma referente à variação dessa frequência de maneira mais clara e concisa.

Podemos perceber essa relação principalmente na segunda parte do experimento onde sinais com uma maior oscilação foram capazes de entregar espectrogramas mais sensíveis e caracteristicos, mostando assim uma variação de frequência nesse sinal ao decorrer do tempo. No mesmo exercício podemos percereber que quando um sinal tem uma menor frequência, a transformada rápida de fourier não se mostra tão eficiente.
